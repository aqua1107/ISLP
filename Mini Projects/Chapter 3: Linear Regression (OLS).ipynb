```python
Boston = load_data("Boston")
Boston.columns
```

```python
X = pd.DataFrame({'intercept': np.ones(Boston.shape[0]),
'lstat': Boston['lstat']})
X[:4]
```

```python
y = Boston['medv']
model = sm.OLS(y, X)
results = model.fit()
```

```python
summarize(results)
```

```python
design = MS(['lstat'])
design = design.fit(Boston)
X = design.transform(Boston)
X[:4]
```
